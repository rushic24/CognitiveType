```python
from font2img.font2img import font2img
from datetime import datetime
now_str = datetime.now().strftime('%Y-%m-%d_%H%M%S')

src_chars_txt_path = 'font2img/src_chars_txt/alphabets_hankaku_caps.txt'

# result save path
font_pngs = 'result/' + now_str
if not os.path.exists(font_pngs):
    os.makedirs(font_pngs)

# upload fonts at font_ttfs
font_ttfs = 'fontttfs'
if not os.path.exists(font_ttfs):
    os.makedirs(font_ttfs)

f2i = font2img(src_font_dir_path=font_ttfs,
                src_chars_txt_path=src_chars_txt_path,
                dst_dir_path=font_pngs, # save at dst_dir_path
                canvas_size=64,
                font_size=0,
                output_ext='png',
                is_center=True,
                is_maximum=False,
                is_binary=False,
                is_unicode=False,
                is_by_char=True,
                is_recursive=True)
f2i.run()
```
