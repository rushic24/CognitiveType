```python
names = list(set(df['font_name'].apply(lambda x: x.lower())))
[x for x in names if  x]
sorted(names)
```

```python
 y11 = pd.Series(df['font_name'].unique()).apply(lambda x: x.lower()).str.split('-').apply(lambda x: x[0]).value_counts()
 y11 = y11[y11 >= 5]
```

```python
y11.index
```

```python
df2 = df[df['font_name'].apply(lambda x: x.lower()).str.split('-').apply(lambda x: x[0]).isin(y11.index)]
print(f"df2shape: {df2.shape}")
df2.head()
```

```python
del df
```

```python
df2 = df2[df2['label'].str.contains("[A-Za-z0-9]+$")].reset_index(drop=True)
df2.head()
```

```python
df2.to_csv('min_5_fonts_english.csv', index=False)
```
